##1. Difference in the average number of bikes rented on weekdays vs. weekends

```python
# 1. Formulate Hypothesis
# Question: Is there a difference in the average number of bikes rented on weekdays vs. weekends?
# H0: The average number of bikes rented is the same on weekdays and weekends.
# H1: The average number of bikes rented is different on weekdays and weekends.
```

```python
# 2. Prepare the Data
# Create a new column 'day_type'
df['day'] = df['Date'].apply(lambda date: 'weekday' if pd.to_datetime(date).weekday() < 5 else 'weekend')
```

```python
df.head()
```

```python
# Separate bike rental counts for weekdays and weekends
weekday_rentals = df[df['day'] == 'weekday']['Rented Bike Count']
weekend_rentals = df[df['day'] == 'weekend']['Rented Bike Count']
```

```python
# 3. Perform the Test (Independent Samples T-test)
from scipy import stats
t_statistic, p_value = stats.ttest_ind(weekday_rentals, weekend_rentals)
#The calculated t-statistic, which measures the difference between the means of the two groups in terms of standard error.
```

```python
# 4. Interpret the Results
alpha = 0.05

print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in bike rentals between weekdays and weekends.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in bike rentals between weekdays and weekends.")
```

##2. Testing the Correlation Between Temperature and Bike Rentals

```python
df.head()
```

```python
temperature = df['Temperature(째C)']
rentals = df['Rented Bike Count']
```

```python
correlation_coefficient, p_value = stats.pearsonr(temperature, rentals)
```

```python
# Print the results
print(f"Correlation Coefficient: {correlation_coefficient:.2f}")
print(f"P-value: {p_value:.3f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant correlation between temperature and bike rentals.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation between temperature and bike rentals.")
```
